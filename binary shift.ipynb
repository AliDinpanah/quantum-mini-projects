```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator


def shift_right(input):
    qc = QuantumCircuit(4)


    for i, bit in enumerate(reversed(input)): 
        if bit == '1':
            qc.x(i) 


    qc.swap(0, 1)
    qc.swap(1, 2)
    qc.swap(2, 3)


    qc.measure_all()

    simulator = AerSimulator()
    compiled_circuit = transpile(qc, simulator)
    job = simulator.run(compiled_circuit)
    result = job.result()


    counts = result.get_counts()
    return counts

def shift_left(input):
    qc = QuantumCircuit(4)


    for i, bit in enumerate(reversed(input)): 
        if bit == '1':
            qc.x(i) 

    qc.swap(2, 3)
    qc.swap(1, 2)
    qc.swap(0, 1)
    
  


    qc.measure_all()

    simulator = AerSimulator()
    compiled_circuit = transpile(qc, simulator)
    job = simulator.run(compiled_circuit)
    result = job.result()


    counts = result.get_counts()
    return counts


input = "1011"
right=shift_right(input)
left= shift_left(input)
print('right shift: \t' + max(right))
print('left shift: \t' + max(left))
```
